# Maksukyvyttömyysennustus-malli

Tarkoituksena rakentaa "tekoäly", machine learning-malli, joka oppii ennustamaan velallisen sortumisen maksukyvyttömyyteen.

Kaiken datan on tarjonnut GAURAV DUTTA, Kaggle.com, ja malli on rakennettu lukio-opiskelijan toimesta oppimiskokemuksena.

```python
# Otetaan hyödyllisiä python-kirjastoja käyttöön

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#malli
from sklearn.ensemble import RandomForestClassifier 
import sklearn.metrics as metrics
malli=RandomForestClassifier()

from sklearn.model_selection import train_test_split

#data

data=pd.read_csv(r'C:\Users\toivo.haapasaari\train_LZV4RXX.csv')

data
```

```python
#yhläällä on data. Nyt lisätään dataan muutama oma tunnusluku,
#jotta parannamme mallia.

#jos poikkeuksellisen suuri osuus omaisuudesa:

std=np.std(data['asset_cost']/data['loan_amount']-1)
mu=np.mean(data['asset_cost']/data['loan_amount']-1)

Z=lambda x,y: ((x/y-1)-mu)/std
data['Check_Laina']=Z(data['asset_cost'],data['loan_amount'])  
```

```python
#valmistellaan malli

data=data.dropna()
X=data.drop(['loan_default','loan_id','proof_submitted'],axis=1)
Y=data['loan_default']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)

malli.fit(X_train,y_train)

```

```python
#arvoidaan mallin tehokkuus

a=malli.score(X_test, y_test)*100
print('malli ennustaa noin:',("%.1f" %a),'% tarkkuudella dataa ,jota se ei ole vielä nähnyt')
```

oppimiskokemuksen tulos on miellyttävä, vaikka 60% tuloksessa on vielä parannettavaa.

Mallia voisi vielä parannella esimerkiksi luokittelemalla vastaajien iät maksukyvyttömyysriskin perusteella, sekä kyselemällä velalliselta lisää heidän sosioekonomisesta tilanteesta. Esimerkiksi tiedosta siitä, että onko velallinen vakituisessa työsuhteessa vai ei olisi hyötyä.
