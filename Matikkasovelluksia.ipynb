```python
def integraali(a,b,n):
    h=(b-a)/n
    r=0
    for k in range(1,n+1):
        f=(a+k*h)**2+1
        r+=f*h
    return r
```

```python
print("i ",integraali(0,1,5))
print("ii ", integraali(-1,2,1000))
```
