```python
import pandas as pd
import pandas_ta as ta
import yfinance as yf
```

```python
tic=yf.Ticker('AAPL')
df=tic.history('max')
```

```python
df=df.drop(['Dividends','Stock Splits'],axis=1)
```

```python
df['RSI']=ta.momentum.rsi(df['Close'])
df.dropna()
```

```python
import matplotlib.pyplot as plt
plt.plot(df.drop(['Volume'],axis=1))
plt.figsize=100
plt.show()
```
